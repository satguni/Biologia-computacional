```python
#taller 1 sergio andres tirado gallego
!pip install pandas matplotlib --quiet  # Instalar las librerías
#titulo: conteo de polen en miel en apiarios cerca de cukltivos de diversas plantas fructiferas y de uso comercial mas importantes para la economia de brasil
#desde la literatura se ha propuesto que la incorporacion de abejas en los cultivos ayuda al aumento de la produccion a mejorar la salud de los cultivos y
#evita la polinizacion manual o ineficaz
import pandas as pd
import matplotlib.pyplot as plt
#se tomaron datos de polen en miel contados manualmente en apiarios de apis mielifera que se ubicaron estrategicamente dentro de plantaciones comerciales
#para asi observar como interactuan las abejas cuando se las pone dentro de u ambiente con disponibilidad de un tipo de polen especifico las plantaciones tenianm
#medidas similares y se coloco las colmenas en puntos centrales y espaciados entre ellos.
# Diccionario con los datos de polen
data = {
    'Familia': [
        'Asteraceae Helianthus',
        'Brassicaceae Brassica',
        'Betulaceae Betula',
        'Fagaceae Quercus',
        'Ericaceae Vaccinium',
        'Apiaceae Daucus',
        'Cucurbitaceae Cucurbita',
        'Lamiaceae Salvia',
        'Amaranthaceae Amaranthus',
        'Sapindaceae Acer',
        'Anacardiaceae Mangifera',
        'Moraceae Ficus',
        'Euphorbiaceae Ricinus',
        'Rutaceae Citrus'
    ],
    'Conteo': [534, 128, 763, 349, 627, 402, 215, 839, 791, 264, 677, 356, 93, 712]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Calcular porcentaje
df['Porcentaje'] = df['Conteo'] / df['Conteo'].sum() * 100

# Ordenar por porcentaje de mayor a menor
df = df.sort_values(by='Porcentaje', ascending=False)

# Mostrar la tabla ordenada
print(df)
```

```python
plt.figure(figsize=(10,6))
plt.barh(df['Familia'], df['Porcentaje'], color='lightgreen')
plt.xlabel('% de polen')
plt.title('Distribución porcentual del polen por familia')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
#pudimos observar que las especies de salvia, amaranthus, betula obtuvieron mayor cantidad de polen contada lo que nos indica que las abejas o visistan mas
#este tipo de plantas o que se cargan mas de polen por la cantidad que liberan estas plantas para plantas como brassica o ricinus tambien se tuvieron conteos
#y aunque mas pocos podemos decir que las abejas las visitan y polinizan sin discriminar.
```
